```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(
    'tmp_train.csv',
    keep_default_na=False
)
df
```

```python
df = df[df['id'] != '>'].reset_index(drop=True)  # need to reset index
df
```

```python
df[df['rank'] == '']
```

```python
df.iloc[6241991, 4] = 46
df.iloc[6241991, -1] = 3
```

```python
""" change dataframe's dtype """
df = df.astype({'rank': 'int', 'fold': 'int'})  # 얘는 이렇게 대입을 시켜줘야
```

```python
df
```

```python
df.to_csv('perfect_tmp_train.csv', index=False)
```

```python
id_list = df.id.unique()
cell_id_list = df.groupby('id')['cell_id'].agg(list)
cell_type_list = df.groupby(['id'])['cell_type'].apply(list)
rank_list = df.groupby(['id'])['rank'].apply(list)
source_list = df.groupby(['id'])['source'].apply(list)
fold_list = df.groupby(['id'])['fold'].apply(list)
len(cell_id_list)
```

```python
new_df = pd.DataFrame()
new_df['id'] = id_list
new_df['cell_id'] = [i for i in cell_id_list]
new_df['cell_type'] = [i for i in cell_type_list]
new_df['rank'] = [i for i in rank_list]
new_df['source'] = [i for i in source_list]
new_df['fold'] = [i[0] for i in fold_list]
```

```python
new_df.to_csv('perfect_final_train.csv', index=False)
```
